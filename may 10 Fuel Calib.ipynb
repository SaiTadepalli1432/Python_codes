```python
# Enetering values
rad = float(input("Enter the radius of the opening of the cyclindrical tank:"))
leng = float(input("Enter the length of the cyclindrical tank:"))
dia = 2* radius

question = print('''If you want to know the full calibration table enter "1".''' '\n'
                 '''If you want to know volume of fuel for a particular height enter "0"''' )  
answer = int(input('Answer:'))

if answer == 0:
    hei = float(input("Enter the height for which you want to know the volume of fuel in the cyclindrical tank:"))
    
else:
    pass 
    
    



# Defining the fucntion calib

import math
 

def calib(radius, height, length):
    h = float(height)
    r = float(radius)
    l = float(length)
    
    if h<r or h==r:
        theta1 = float(2*(math.acos(1-(h/r))))
    
        # area of minor segment
    
        area_seg = float((r**2)/2) * ((theta1) - math.sin(theta1))
   
    
        volume = area_seg * l
        return volume
        
    else:
        theta2 = float(2*(math.acos((h/r)-1)))
    
        # area of major segment
    

        area_upper_seg = float((r**2)/2)* ((theta2) - math.sin(theta2))
       
        area_circle = math.pi * (r**2)
        
        total_area = area_circle  - area_upper_seg
   
    
        volume = total_area * l
        
        return volume
        
    
    
    
# Aligning according to 4096 points on the censor

num_parts = 4096
step = diameter/4096
i =0
dictionary = {}

for x in range((num_parts+1)):
    dictionary[x] = i
    i += step
    
    # print (dictionary)
    
    
    
    
    
# Getting the index and the volume of fuel for each index
list_values = [v for v in dictionary.values()]
list_keys = [k for k in dictionary.keys()]

volume_list = []

for v in range(len(list_values)):
    volume_list.append(calib(rad,v,leng)) 
    
    
    
#Merging the two lists to get tuples(index, volume of fuel at index)
merged_list = tuple(zip(list_keys, volume_list)) 
merged_list 
    
    
```
